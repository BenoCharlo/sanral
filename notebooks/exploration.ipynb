## **DATE-RELATED FE**

```python
weekday_freq = train.name_day.value_counts(True).reset_index()
weekday_freq.columns = ["name_day", "days_freq"]

month_freq = train.Month.value_counts(True).reset_index()
month_freq.columns = ["month", "month_freq"]

day_freq = train.Day.value_counts(True).reset_index()
day_freq.columns = ["day", "day_freq"]

hour_freq = train.Hour.value_counts(True).reset_index()
hour_freq.columns = ["hour", "hour_freq"]
```

```python
new_data = new_data.merge(weekday_freq, on="name_day", how="left")

new_data = new_data.merge(month_freq, on="month", how="left")

new_data = new_data.merge(day_freq, on="day", how="left")

new_data = new_data.merge(hour_freq, on="hour", how="left")
```

# **EXPLORATORY DESCRIPTIVE ANALYSIS**

```python
train["Occurrence Local Date Time"] = pd.to_datetime(train["Occurrence Local Date Time"])
train["datetime"] = [datetime.datetime.strptime(str(date).split(" ")[0], "%Y-%m-%d") for date in train["Occurrence Local Date Time"]]
train["Year"] = [date.year for date in train["Occurrence Local Date Time"]]
train["Month"] = [date.month for date in train["Occurrence Local Date Time"]]
train["Day"] =[date.day for date in train["Occurrence Local Date Time"]]
train["Hour"] = [date.hour for date in train["Occurrence Local Date Time"]]
train["name_day"] = [date.strftime("%A") for date in train["Occurrence Local Date Time"]]
```

```python
train.head()
```

```python
train["road_segment_id"].value_counts(True).reset_index()
```

```python
daily_counts = train.groupby(["datetime"]).size().reset_index()
daily_counts.columns = ["datetime", "count"]
```

```python
daily_counts["MA_5"] = daily_counts["count"].rolling(60, center=True).mean()
```

```python
daily_counts
```

```python
plt.figure(figsize=(18, 9))
sns.lineplot(x="datetime", y="MA_5", data=daily_counts, color="red")
sns.lineplot(x="datetime", y="count", data=daily_counts)
plt.xticks(rotation=45)
```

```python
monthly_counts = train.groupby(["Year", "Month"]).size().reset_index()
monthly_counts.columns = ["Year", "Month", "count"]
monthly_counts["datetime"] = ["-".join([str(x),str(y)]) for x, y in zip(monthly_counts.Year, monthly_counts.Month)]
monthly_counts["datetime"] = [datetime.datetime.strptime(str(date), "%Y-%m") for date in monthly_counts["datetime"]]
```

```python
monthly_counts.head(25)
```

```python
plt.figure(figsize=(18, 9))
sns.lineplot(x="datetime", y="count", data=monthly_counts)
plt.xticks(rotation=45)
```

```python
day_counts = train.groupby(["name_day"]).size().reset_index()
day_counts.columns = ["name_day", "count"]
day_counts = day_counts.sort_values("count", ascending=False)
```

```python
plt.figure(figsize=(12, 9))
sns.barplot(x="count", y="name_day", data=day_counts)
plt.xticks(rotation=45)
```
